# EQ Ranking Analysis Notebook
This notebook analyzes EQ ranking experiment data.

Input format:
```python
{
  "SubjectName": [rankEQ1, rankEQ2, rankEQ3, ...]
}
```

- List index = EQ index
- Value = ranking (1 = best)

The notebook will compute:
1. EQ Mean Rank
2. Rank Standard Deviation
3. Top4 / Bottom4 EQ per subject


````python
import numpy as np
import pandas as pd
````

## Input Scores
Edit this section to add more subjects or change rankings.

````python
scores = {
    "Parker":  [8,5,2,9,3,6,1,4,11,12,13,14,7,15,16,17,10],
    "Ken":    [4,1,8,2,7,14,3,6,9,5,10,11,12,16,13,15,17],
    "Roger":  [7,1,5,11,3,6,2,9,4,10,12,13,8,15,17,16,14],
    "David":  [10,1,12,3,2,13,4,8,5,11,7,9,6,15,16,14,17]
}

EQ_NAMES = [f"EQ{i+1}" for i in range(17)]
````

## Convert to DataFrame

````python
df = pd.DataFrame(scores)
df.index = EQ_NAMES

print("Ranking Table")
display(df)
````

## EQ Mean Rank and Ranking Variance

````python
analysis = pd.DataFrame()

analysis["Mean_Rank"] = df.mean(axis=1)
analysis["Rank_SD"] = df.std(axis=1)

analysis = analysis.sort_values("Mean_Rank")

print("EQ Ranking Analysis")
display(analysis)
````

## Top4 / Bottom4 EQ for Each Subject

````python
result = {}

for subject in df.columns:
    ranking = df[subject]
    
    top4 = ranking.nsmallest(4)
    bottom4 = ranking.nlargest(4)
    
    result[subject] = {
        "Top4": list(top4.index),
        "Bottom4": list(bottom4.index)
    }

result
````

## Table View of Subject Preferences

````python
rows = []
    
for subject in df.columns:
    ranking = df[subject]
    
    top4 = list(ranking.nsmallest(4).index)
    bottom4 = list(ranking.nlargest(4).index)
    
    rows.append({
        "Subject": subject,
        "Top4": ", ".join(top4),
        "Bottom4": ", ".join(bottom4)
    })

subject_pref = pd.DataFrame(rows)

display(subject_pref)
````
